## 2. Dataset Description
The dataset contains information on individualsâ€™ age, sex, BMI, number of children, smoker status, region, and medical charges.


```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("insurance.csv")
df.head()
df.info()
df.describe()

```

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['sex'] = le.fit_transform(df['sex'])
df['smoker'] = le.fit_transform(df['smoker'])
df['region'] = le.fit_transform(df['region'])
```

## 3. Visualizations & Insights


```
sns.boxplot(x='smoker', y='charges', data=df)
plt.title("Charges by Smoker Status")
```

```
sns.scatterplot(x='bmi', y='charges', hue='smoker', data=df)
plt.title("BMI vs Charges")
```

```
sns.barplot(x='region', y='charges', data=df)
plt.title("Average Charges by Region")
```

```
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
```

### 1. Charges by Smoker Status

The boxplot shows a significant difference in charges between smokers and non-smokers.
- **Smokers**: Higher charges, with a wide range, indicating that smokers have more extreme medical costs.
- **Non-smokers**: More consistent, lower charges on average.
  
**Conclusion**: Smoking is a key factor that increases healthcare costs.


```
sns.boxplot(x='smoker', y='charges', data=df)
plt.title("Charges by Smoker Status")
```

### 2. BMI vs Charges (Colored by Smoker Status)

The scatter plot reveals that **higher BMI** is associated with **higher medical charges**.
- **Smokers**: Those with high BMI have the **highest charges**, which indicates that smoking exacerbates the effect of BMI on healthcare costs.
- **Non-smokers**: A similar trend is observed, but the charges are generally lower for non-smokers.

**Conclusion**: Both high BMI and smoking contribute significantly to higher medical costs.


```
sns.scatterplot(x='bmi', y='charges', hue='smoker', data=df)
plt.title("BMI vs Charges")
```

### 3. Average Charges by Region

The bar plot indicates slight differences in average medical charges across regions.
- The **Southeast** region has the **highest average charges**. This could be due to factors like regional healthcare accessibility or lifestyle choices.
- Other regions show similar, lower charges on average.

**Conclusion**: Regional factors may play a role in healthcare costs, with the Southeast region showing the highest charges.


```
sns.barplot(x='region', y='charges', data=df)
plt.title("Average Charges by Region")
```

### 4. Correlation Matrix

The heatmap highlights key relationships between different features and charges.
- **Smoker status**: Shows the strongest positive correlation with **charges** (correlation = 0.79).
- **BMI**: Also has a moderate positive correlation (0.19) with **charges**.
- **Age** and **children** show weaker correlations with **charges**.

**Conclusion**: The strongest predictor of medical charges in this dataset is whether or not the person smokes. BMI also plays a role, but less significantly.


```
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
```

## Conclusion
This analysis shows that smoking and BMI are the primary factors driving healthcare costs. Interventions aimed at reducing smoking and managing BMI could significantly reduce healthcare expenses.
